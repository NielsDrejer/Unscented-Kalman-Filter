# Notebook for plotting NIS values

```python
import matplotlib.pyplot as plt
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 15, 10

nis_radar = []
nis_laser = []

with open('./build/nis.txt') as f:
    for line in f:
        line = line.rstrip('\n')
        if line[:9] == "NIS Radar":
            nis_radar.append(line.split('=')[-1].strip())
        else:
            nis_laser.append(line.split('=')[-1].strip())      
```

```python
plt.plot(nis_radar[1:])
plt.axhline(y=7.815, xmin=0, xmax=1, linewidth=1, color='r')
plt.title('NIS for Radar')
plt.ylabel('NIS value')
plt.xlabel('Measurement')
plt.show()
```

```python
plt.plot(nis_laser)
plt.axhline(y=5.991, xmin=0, xmax=1, linewidth=1, color='r')
plt.title('NIS for Laser')
plt.ylabel('NIS value')
plt.xlabel('Measurement')
plt.show()
```
